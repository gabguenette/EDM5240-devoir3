# Analyse des demandes d'accès à l'information canadienne depuis 2015

Le fichier ati.csv contient les demandes d'accès à l'information déposées aux ministères et agences du gouvernement canadien depuis 2015. Plusieurs faits intéressants peuvent être constatés lorsque nous analysons ce fichier. 

```python
import pandas as pan
```

```python
info = pan.read_csv("ati.csv")
```

```python
info
```

Premièrement, le fichier nous apprend que le ministère des Affaires mondiales du Canada est celui ayant reçu le plus de demandes d'accès à l'information depuis 2015, avec un total de 1799 demandes. Il est suivi de près par le ministère de la Défense nationale et par celui de l'Emploi et du développement social, à 1703 et 1538 demandes respectivement. 

```python
info.titreOrganisation.value_counts()
```

Le ministère des Affaires mondiales du Canada totalise 6.36% des demandes d'accès à l'information.

```python
info.titreOrganisation.value_counts() / len(info.titreOrganisation)*100
```

Deuxièmement, certaines demandes n'ont pas donné de résultats concluants. Ainsi, aucun document n'a été créé. En effet, il y a 5 807 demandes d'accès à l'information qui n'ont aucun document associé (représenté par un nombre de pages égal à 0). Le tout représente tout de même une demande sur cinq (20,52%).

```python
nodoc = info.pages == 0
```

```python
nodoc.value_counts()
```

```python
(5807/(5807+22487))*100
```

Finalement, malgré le fait que ce soit le ministère des Affaires mondiales du Canada qui a reçu le plus de demandes, il est intéressant de constater qu'Élections Canada a remis le plus grand nombre de pages en demandes d'accès à l'information, avec un total de 1 704 954 pages. De son côté, le ministère des Affaires mondiales du Canada en a remis 132 540.
